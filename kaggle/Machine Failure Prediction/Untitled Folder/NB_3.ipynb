```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
# 不要な警告文非表示
warnings.filterwarnings("ignore")
from pycaret.classification import *
```

```python
train = pd.read_csv('train.csv', index_col=0)
bulk = pd.read_csv('machine failure.csv')
bulk = bulk.drop(columns=['UDI'])
train_df = pd.concat([train, bulk], axis=0)
test_df = pd.read_csv("test.csv", index_col=0)
```

```python
X = train_df.drop(columns=['Product ID', 'Machine failure'])
y = train_df['Machine failure']
```

```python
exp1 = setup(train_df, target = 'Machine failure', index=False)
```

```python
compare_models()
```

```python
model = create_model('nb')
```

```python
tuned_model = tune_model(model, optimize = 'AUC', round=10, early_stopping=True, verbose=True)
```

```python
evaluate_model(tuned_model)
```

```python
final_model = finalize_model(tuned_model)
```

```python
pred = predict_model(final_model, data = test_df)
```

```python
pred_final = final_model.predict_proba(test_df)[:, 1]
```

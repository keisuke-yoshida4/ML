```python
import warnings
# 不要な警告文非表示
warnings.filterwarnings("ignore")
from pycaret.classification import *
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.DataFrame(iris.target, columns=["target"])
df = pd.concat([X,y], axis=1)
```

```python
df
```

```python
# setupで欠損値処理、データ分割を自動で行ってくれる　
# 第一引数：解析に用いる data（読み込んだデータ）,第二引数：予測に用いる目的変数の名称
exp1 = setup(df, target = 'target')
```

```python
# 学習モデルによる精度の比較
compare_models()
```

```python
# モデリング
model = create_model('dt')
```

```python
# ハイパーパラメータの最適化
# 第一引数：モデルの名称,第二引数：最適化の対象
tuned_model = tune_model(model, optimize = 'Accuracy')
```

```python
print(model)
```

```python
print(tuned_model)
```

```python
# モデルの評価指標を確認（解析結果の可視化）
evaluate_model(tuned_model)
```

```python
# 作成したモデルの読み込み
final_model = finalize_model(tuned_model)
```

```python
# 予測
pred = predict_model(final_model, data = X)
```
